```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
breast_cancer = load_breast_cancer()
breast_cancers_data = breast_cancer.data
breast_cancers_label = breast_cancer.target
X_train, X_test, y_train, y_test = train_test_split(breast_cancers_data, 
                                                    breast_cancers_label, 
                                                    test_size=0.2, 
                                                    random_state=7)
print(breast_cancers_data.shape)
```

```python
decision_tree = DecisionTreeClassifier(random_state=11)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
Random_Forest = RandomForestClassifier(random_state=135)
Random_Forest.fit(X_train, y_train)
y_pred_rf = Random_Forest.predict(X_test)

print(classification_report(y_test, y_pred_rf))
```

```python
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print(classification_report(y_test, y_pred_svm))
```

```python
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred_sgd = sgd_model.predict(X_test)

print(classification_report(y_test, y_pred_sgd))
```

```python
logistic_model = LogisticRegression(max_iter=10000)
logistic_model.fit(X_train, y_train)
y_pred_log = logistic_model.predict(X_test)

print(classification_report(y_test, y_pred_log))
```

유방암 Recall, F1-score    
유방암 데이터의 경우는 Recall, F1-score가 중요한 평가지표이다.  
양성인 환자를 음성으로 판별하는 일이 적어야하기 때문에  
Recall값을 우선으로 두어야 하고  
양성 데이터와 음성 데이터간의 갯수 차이가 크기 때문에  
F1-score도 중요하게 생각해야 한다.  
RandomForestClassifier, LogisticRegression을 사용하는 것이 좋아보인다.
