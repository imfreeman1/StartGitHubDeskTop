```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
import dlib
from google.colab import drive
drive.mount('/content/drive')

print("üå´üõ∏")
```

```python
my_image_path = '/content/drive/MyDrive/·Ñå·Ö¶·ÑÜ·Ö©·Ü®·Ñã·Ö•·Üπ·ÑÇ·Ö≥·Ü´ ·Ñë·Ö©·ÜØ·ÑÉ·Ö•/image2.png'
img_bgr = cv2.imread(my_image_path)    # OpenCVÎ°ú Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§
img_show = img_bgr.copy()      # Ï∂úÎ†•Ïö© Ïù¥ÎØ∏ÏßÄÎ•º Îî∞Î°ú Î≥¥Í¥ÄÌï©ÎãàÎã§
plt.imshow(img_bgr)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python
detector_hog = dlib.get_frontal_face_detector()
print("üå´üõ∏")
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 2)   # (image, num of image pyramid)
print("üå´üõ∏")
```

```python
print(dlib_rects)   

for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA)

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = '/content/drive/MyDrive/·Ñå·Ö¶·ÑÜ·Ö©·Ü®·Ñã·Ö•·Üπ·ÑÇ·Ö≥·Ü´ ·Ñë·Ö©·ÜØ·ÑÉ·Ö•/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
print("üå´üõ∏")
```

```python
list_landmarks = []

# ÏñºÍµ¥ ÏòÅÏó≠ Î∞ïÏä§ ÎßàÎã§ face landmarkÎ•º Ï∞æÏïÑÎÉÖÎãàÎã§
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    # face landmark Ï¢åÌëúÎ•º Ï†ÄÏû•Ìï¥Îë°ÎãàÎã§
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
    list_landmarks.append(list_points)

print(len(list_landmarks[0]))
```

```python
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[30]) # ÏΩîÏùò indexÎäî 30 ÏûÖÎãàÎã§
    x = landmark[30][0]
    y = landmark[30][1] - dlib_rect.height()//2
    w = h = dlib_rect.width()
    print ('(x,y) : (%d,%d)'%(x,y))
    print ('(w,h) : (%d,%d)'%(w,h))
```

```python
sticker_path = '/content/drive/MyDrive/·Ñå·Ö¶·ÑÜ·Ö©·Ü®·Ñã·Ö•·Üπ·ÑÇ·Ö≥·Ü´ ·Ñë·Ö©·ÜØ·ÑÉ·Ö•/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path) # Ïä§Ìã∞Ïª§ Ïù¥ÎØ∏ÏßÄÎ•º Î∂àÎü¨ÏòµÎãàÎã§
img_sticker = cv2.resize(img_sticker, (w,h))
out = img_sticker.copy()
out = 255 - out
print (out.shape)
# print(img_sticker.shape)
```

```python
refined_x = (x - w // 2) - 16
refined_y = y
print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
if refined_x < 0: 
    img_sticker = img_sticker[:, -refined_x:]
    refined_x = 0
if refined_y < 0:
    img_sticker = img_sticker[-refined_y:, :]
    refined_y = 0

print ('(x,y) : (%d,%d)'%(refined_x, refined_y))
```

```python
sticker_area = img_show[refined_y:refined_y+out.shape[0], refined_x:refined_x+out.shape[1]]
img_show[refined_y:refined_y+out.shape[0], refined_x:refined_x+out.shape[1]] = \
    np.where(out==0,sticker_area,out).astype(np.uint8)
print("Ïäù~")
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show()
```
